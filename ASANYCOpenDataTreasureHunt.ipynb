#Question 1: Getting first 100,000 yellow taxi trips in 2023
##First just making sure everything works by pulling the first 5 records, as 100k records may take a little while

```python
from sodapy import Socrata

# Your App Token can be found at https://data.cityofnewyork.us/profile/edit/api_credentials
# For public datasets, an app token is not strictly required, but it's good practice.
# If you have an app token, replace 'YOUR_APP_TOKEN' with your actual token.
app_token = None # Replace with your app token if available

# Domain and dataset identifier from the URL
domain = 'data.cityofnewyork.us'
dataset_identifier = '4b4i-vvec'

# Initialize the Socrata client
client = Socrata(domain, app_token)

# Fetch the first 5 records to verify the connection
print(f"Fetching the first 5 records from {dataset_identifier}...")
results = client.get(dataset_identifier, limit=5)


# Close the client connection
client.close()
```

##Now pulling the first 100,000 records with a fare > 0 and collecting them

```python
soql_where_clause = "fare_amount > '0'"

print(f"Fetching records with a limit of 100000...")
first100kData = client.get(dataset_identifier, where=soql_where_clause, limit=100000)

print(f"Fetched {len(first100kData)} records.")

fare_amounts_100k = [float(record['fare_amount']) for record in first100kData if 'fare_amount' in record]
```

#Question 3: Importing the Motor Vehicle Collisions – Crashes dataset
##Update dataset_id and pull in data.


## Fetch Relevant Collision Data for Staten Island in 2024



```python
dataset_identifier = 'h9gi-nx95' # Corrected dataset identifier
records_to_fetch = 10000 #There are fewer than 10000 records of interest to us based on our query
soql_query_street_names = "on_street_name IS NOT NULL AND crash_date >= '2024-01-01T00:00:00' AND crash_date < '2025-01-01T00:00:00' AND borough = 'STATEN ISLAND'"

print(f"Fetching up to {records_to_fetch} records from {dataset_identifier} with street names and no nulls...")
collision_data = client.get(
    dataset_identifier,
    select='on_street_name, borough, crash_date', # Included cross_street_name
    where=soql_query_street_names,
    limit=records_to_fetch
)

print(f"Fetched {len(collision_data)} records for street name analysis.")

# Display the first few records to verify the data structure
# for i, record in enumerate(collision_data[:5]):
#     print(f"Record {i+1}: {record}")
```

#Question 4: Importing the 311 Service Requests from 2020 to Present and the Open Streets Locations datasets


##First let's get the Open Streets Locations Dataset

```python
# Define the dataset identifier and limit for the
dataset_id = 'uiay-nctu'
limit_count = 5000

print(f"Fetching Open Streets Locations...")

# Fetch the data using the initialized Socrata client
open_streets_data = client.get(dataset_id, limit=limit_count)

# Print the result count
print(f"Successfully fetched {len(open_streets_data)} records.")
```

##Next let's fetch data for Cooper Street or Seaman Avenue in 10034, though we will need to do more filtering based on Open Streets info

```python
# Define the SoQL where clause based on the instructions
soql_where = (
    "(street_name = 'COOPER STREET' OR street_name = 'SEAMAN AVENUE') "
    "AND incident_zip = '10034' "
)

# Define the dataset identifier and limit for the
dataset_id = 'erm2-nwe9'
limit_count = 50000

print(f"Fetching requests for COOPER STREET and SEAMAN AVENUE ")

# Fetch the data using the initialized Socrata client
requests_data = client.get(dataset_id, where=soql_where, limit=limit_count)

# Print the result count
print(f"Successfully fetched {len(requests_data)} records.")
```
